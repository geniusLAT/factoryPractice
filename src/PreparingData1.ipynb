Хочу создать датафрейм, который объединит нормальную таблицу и данные геолокации. При этом согласно плану удаляем столбик geo_id, как не представляющий интерес.

```python
train_df_with_geo = train_csv.merge(geo_info_csv, on='geo_id', how='left').drop(columns = ['geo_id'])
test_df_with_geo = test_csv.merge(geo_info_csv, on='geo_id', how='left').drop(columns = ['geo_id'])

train_df_with_geo
```

Аналогично обрабатываем referer_vectors

```python
train_df_with_referer = train_df_with_geo.merge(referer_vectors_csv, on='referer', how='left').drop(columns = ['referer'])
test_df_with_referer = test_df_with_geo.merge(referer_vectors_csv, on='referer', how='left').drop(columns = ['referer'])

train_df_with_referer
```

Далее нам потребуется распарсить _user_agent_ 

```python
import json
import versionsOS

browsers = []

ok_versions = []
wrong_versions = []

ok_count = 0
not_ok_count = 0

def parse(json_string):
    global ok_count
    global not_ok_count
    json_string = str(json_string)

    # Заменяем одинарные кавычки на двойные
    
    json_string = json_string.replace("'", '"')

    try:
        # Преобразуем строку в словарь
        data = json.loads(json_string)

        # Создаем массив из ключей и значений
        arr = []
        for key, value in data.items():
            arr.append(value)
        
        if len(arr) !=4:
            return ["wrong length ", "0", " ", "0"]

        versionOS = versionsOS.get_os_release_year(arr[2], arr[3])
        #print(versionOS)

        if versionOS == None:
            not_ok_count= not_ok_count +1
            versionOS = "None: " +arr[3]
            os_and_version = [arr[2], arr[3]]
            if not wrong_versions.__contains__(os_and_version):
                wrong_versions.append(os_and_version)
        else:
            ok_count= ok_count + 1
            os_and_version = [arr[2], arr[3]]
            if not ok_versions.__contains__(os_and_version):
                ok_versions.append(os_and_version)

        if not browsers.__contains__( arr[0]):
            browsers.append(arr[0])

        result = [arr[0], arr[1], arr[2], versionOS]

        return result
    except Exception as e:
        return [str(e), "0", " ", "0"]
```

```python
train_parsed_user_agent = train_df_with_referer['user_agent'].apply(parse)
test_parsed_user_agent = test_df_with_referer['user_agent'].apply(parse)

print(browsers)

total = ok_count+not_ok_count
print(total)
print("total:"+str(total)+" ok: "+str(ok_count) +" ("+str(ok_count/total*100)+"%)  "+" not ok: "+str(not_ok_count) +" ("+str(not_ok_count/total*100)+"%)  ")

print(len(wrong_versions))
print(wrong_versions)
print(len(ok_versions))
print(ok_versions)

train_parsed_user_agent
```

Теперь присоединяем новые поля к датафреймам и смотрим на них

```python
train_df_with_referer[['browser', 'browser_version', 'os', 'os_version']] = train_parsed_user_agent.apply(pd.Series)
test_df_with_referer[['browser', 'browser_version', 'os', 'os_version']] = test_parsed_user_agent.apply(pd.Series)

train_df_with_referer
```

Теперь можно убрать из них сам user_agent

```python
train_df_to_save = train_df_with_referer.drop(columns = ['user_agent'])
test_df_to_save = test_df_with_referer.drop(columns = ['user_agent'])

train_df_to_save
```
