Хочу создать датафрейм, который объединит нормальную таблицу и данные геолокации. При этом согласно плану удаляем столбик geo_id, как не представляющий интерес.

```python
train_df_with_geo = train_csv.merge(geo_info_csv, on='geo_id', how='left').drop(columns = ['geo_id'])
test_df_with_geo = test_csv.merge(geo_info_csv, on='geo_id', how='left').drop(columns = ['geo_id'])

train_df_with_geo
```

Аналогично обрабатываем referer_vectors

```python
train_df_with_referer = train_df_with_geo.merge(referer_vectors_csv, on='referer', how='left').drop(columns = ['referer'])
test_df_with_referer = test_df_with_geo.merge(referer_vectors_csv, on='referer', how='left').drop(columns = ['referer'])

train_df_with_referer
```
